```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
one_gold = pd.read_excel(r'C:\Users\dautermanb1.ADWS\Documents\Python Scripts\one year gold.xlsx', usecols = ['Date','Last', '% Change'])
one_crude_oil = pd.read_excel(r'C:\Users\dautermanb1.ADWS\Documents\Python Scripts\one year crude oil.xlsx', usecols = ['Date', 'Last', '% Change'])
one_coffee = pd.read_excel(r'C:\Users\dautermanb1.ADWS\Documents\Python Scripts\one year coffee.xlsx', usecols = ['Date', 'Last', '% Change'])
one_wheat = pd.read_excel(r'C:\Users\dautermanb1.ADWS\Documents\Python Scripts\one year wheat.xlsx', usecols = ['Date', 'Last', '% Change'])
```

```python
one_gold['% Change'].plot(label='Gold', figsize=(32,4),title='Daily % Change over Past Year', color='darkred')
one_crude_oil['% Change'].plot(label='Crude Oil', color='lightgrey')
one_coffee['% Change'].plot(label='Coffee', color='darkblue')
one_wheat['% Change'].plot(label='Wheat', color='darkgreen')
plt.xlabel('Days')
plt.ylabel('% Change')
plt.legend()
```
